```python
#defining the stemming function
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
df = pd.read_csv('Coursera.csv')

# Check for missing values and clean the data
df['Course Description'] = df['Course Description'].fillna('')
df['Skills'] = df['Skills'].fillna('')

# Combine relevant columns for recommendation (Course Description and Skills)
df['Content'] = df['Course Description'] + ' ' + df['Skills']

# Vectorize the content using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Content'])

# Compute cosine similarity between courses
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create a function to get course recommendations
def recommend_courses(course_name, num_recommendations=5):
    # Find the index of the course
    course_idx = df[df['Course Name'].str.contains(course_name, case=False)].index[0]

    # Get similarity scores for the selected course
    sim_scores = list(enumerate(cosine_sim[course_idx]))

    # Sort courses based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the most similar courses
    sim_indices = [i[0] for i in sim_scores[1:num_recommendations+1]]

    # Return the recommended courses
    recommended_courses = df.iloc[sim_indices][['Course Name', 'University', 'Course Rating', 'Course URL']]
    return recommended_courses

# Example usage: Searching for courses related to "Finance for Managers"
search_course = "Finance for Managers"
recommended_courses = recommend_courses(search_course, 5)
print("Recommended courses for:", search_course)
print(recommended_courses)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
df = pd.read_csv('Coursera.csv')

# Check for missing values and clean the data
df['Course Description'] = df['Course Description'].fillna('')
df['Skills'] = df['Skills'].fillna('')

# Combine relevant columns for recommendation (Course Description and Skills)
df['Content'] = df['Course Description'] + ' ' + df['Skills']

# Vectorize the content using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Content'])

# Compute cosine similarity between courses
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to recommend courses based on user's search query
def recommend_courses(course_name, num_recommendations=5):
    # Filter to check if any course matches the input course name
    matching_courses = df[df['Course Name'].str.contains(course_name, case=False)]

    # If no course is found, return a message
    if matching_courses.empty:
        return "Course not found. Please try again with a different name."

    # Get the index of the first matching course
    course_idx = matching_courses.index[0]

    # Get similarity scores for the selected course
    sim_scores = list(enumerate(cosine_sim[course_idx]))

    # Sort courses based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the most similar courses (excluding the selected course itself)
    sim_indices = [i[0] for i in sim_scores[1:num_recommendations+1]]

    # Return the recommended courses
    recommended_courses = df.iloc[sim_indices][['Course Name', 'University', 'Course Rating', 'Course URL']]
    return recommended_courses

# Get user input for the course name
user_input = input("Enter the course name you are searching for: ")

# Get the recommended courses
recommended_courses = recommend_courses(user_input, 5)

# Display the recommended courses
if isinstance(recommended_courses, str):
    print(recommended_courses)
else:
    print("Recommended courses for:", user_input)
    print(recommended_courses)
```
